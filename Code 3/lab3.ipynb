```python
import requests

url = "https://api.football-data.org/v4/competitions/"
response = requests.get(url)
response = response.json()
print(response)
```

```python
import pandas as pd
```

```python
import json

# The API response is a JSON object with a 'competitions' key containing the list of competitions.
# We need to parse the JSON and then extract the 'competitions' list to create a DataFrame.
# Using response.json() is more robust than json.loads(response.text) as it handles character encoding.

data = response
df = pd.DataFrame(data['competitions'])
```
